Q1.Write a Python program that defines classes for 'Student', 'Course', and 'Department'. Implement functionality to enroll a student in a course of a particular department.


```python
class Student:
    def __init__(self, student_id, name):
        self.student_id = student_id
        self.name = name
        self.courses = []

    def __str__(self):
        return f"Student ID: {self.student_id}, Name: {self.name}"

class Course:
    def __init__(self, course_id, course_name):
        self.course_id = course_id
        self.course_name = course_name

    def __str__(self):
        return f"Course ID: {self.course_id}, Name: {self.course_name}"

class Department:
    def __init__(self, department_name):
        self.department_name = department_name
        self.courses_offered = []
        self.enrolled_students = {}

    def add_course(self, course):
        if course not in self.courses_offered:
            self.courses_offered.append(course)
            print(f"Course '{course.course_name}' added to {self.department_name}.")
        else:
            print(f"Course '{course.course_name}' is already offered in {self.department_name}.")

    def enroll_student_in_course(self, student, course):
        if course not in self.courses_offered:
            print(f"Error: Course '{course.course_name}' is not offered by {self.department_name}.")
            return

        if student.student_id not in self.enrolled_students:
            self.enrolled_students[student.student_id] = {'student': student, 'enrolled_courses': []}

        if course not in self.enrolled_students[student.student_id]['enrolled_courses']:
            self.enrolled_students[student.student_id]['enrolled_courses'].append(course)
            student.courses.append(course) # Also add course to student's personal list
            print(f"Student {student.name} ({student.student_id}) enrolled in '{course.course_name}' in {self.department_name}.")
        else:
            print(f"Student {student.name} ({student.student_id}) is already enrolled in '{course.course_name}'.")

    def list_enrollments(self):
        print(f"\nEnrollments for {self.department_name} Department:")
        if not self.enrolled_students:
            print("No students enrolled yet.")
            return
        for student_id, data in self.enrolled_students.items():
            student = data['student']
            enrolled_courses = data['enrolled_courses']
            course_names = [c.course_name for c in enrolled_courses]
            print(f"  {student.name} ({student.student_id}) is enrolled in: {', '.join(course_names) if course_names else 'No courses'}")

# 1. Create instances of Student, Course, and Department
student1 = Student("S001", "Alice Smith")
student2 = Student("S002", "Bob Johnson")
student3 = Student("S003", "Charlie Brown")

course1 = Course("C101", "Introduction to Python")
course2 = Course("C102", "Data Structures")
course3 = Course("C201", "Machine Learning")
course4 = Course("C202", "Database Management")

cs_dept = Department("Computer Science")
math_dept = Department("Mathematics")

# 2. Add courses to departments
print("--- Adding Courses ---")
cs_dept.add_course(course1)
cs_dept.add_course(course2)
cs_dept.add_course(course3)
math_dept.add_course(course4)

# Try adding an existing course (should show message)
cs_dept.add_course(course1)

# 3. Demonstrate enrollment process
print("\n--- Enrollment Process ---")
cs_dept.enroll_student_in_course(student1, course1)
cs_dept.enroll_student_in_course(student1, course2)
cs_dept.enroll_student_in_course(student2, course1)
cs_dept.enroll_student_in_course(student3, course3)

# Try enrolling a student in a course not offered by the department
cs_dept.enroll_student_in_course(student1, course4)

# Try re-enrolling a student in the same course (should show message)
cs_dept.enroll_student_in_course(student1, course1)

# 4. Verify functionality by printing enrolled students and their courses
cs_dept.list_enrollments()
math_dept.list_enrollments()

print(f"\nCourses for {student1.name}: {[c.course_name for c in student1.courses]}")
print(f"Courses for {student2.name}: {[c.course_name for c in student2.courses]}")
print(f"Courses for {student3.name}: {[c.course_name for c in student3.courses]}")
```

Q2.Develop a Python program with a base class 'Bill' and use inheritance to model payment options via 'ChequePayment' and 'CashPayment' classes.


```python
class Bill:
    def __init__(self, bill_id, amount):
        self.bill_id = bill_id
        self.amount = amount

    def display_bill_details(self):
        print(f"Bill ID: {self.bill_id}")
        print(f"Amount: ${self.amount:.2f}")

class ChequePayment(Bill):
    def __init__(self, bill_id, amount, cheque_number):
        super().__init__(bill_id, amount)
        self.cheque_number = cheque_number

    def display_bill_details(self):
        super().display_bill_details()
        print(f"Payment Method: Cheque (Number: {self.cheque_number})")

class CashPayment(Bill):
    def __init__(self, bill_id, amount, tendered_amount):
        super().__init__(bill_id, amount)
        self.tendered_amount = tendered_amount

    def display_bill_details(self):
        super().display_bill_details()
        print(f"Payment Method: Cash")
        print(f"Tendered Amount: ${self.tendered_amount:.2f}")
        change_due = self.tendered_amount - self.amount
        if change_due >= 0:
            print(f"Change Due: Rs{change_due:.2f}")
        else:
            print(f"Amount still owed: Rs{-change_due:.2f}")

print("\n--- Cheque Payment Details ---")
cheque_bill = ChequePayment("B001", 150.75, "CK987654")
cheque_bill.display_bill_details()

print("\n--- Cash Payment Details (exact) ---")
cash_bill_exact = CashPayment("B002", 50.00, 50.00)
cash_bill_exact.display_bill_details()

print("\n--- Cash Payment Details (with change) ---")
cash_bill_change = CashPayment("B003", 75.20, 100.00)
cash_bill_change.display_bill_details()

print("\n--- Cash Payment Details (underpayment) ---")
cash_bill_under = CashPayment("B004", 120.00, 100.00)
cash_bill_under.display_bill_details()
```

Q3. Create a Python class 'Distance' and overload the subtraction operator (-) to subtract two 'Distance' objects.


```python
class Distance:
    def __init__(self, value):
        if not isinstance(value, (int, float)):
            raise TypeError("Distance value must be a number")
        self.value = value

    def __sub__(self, other):
        if not isinstance(other, Distance):
            raise TypeError("Can only subtract another Distance object")
        return Distance(self.value - other.value)

    def __str__(self):
        return f"Distance: {self.value} units"

    def __repr__(self):
        return f"Distance({self.value})"

print("--- Demonstrating Distance Class with Subtraction Operator Overloading ---")

# 1. Create instances of the Distance class
distance1 = Distance(100)
distance2 = Distance(35.5)
distance3 = Distance(120)
distance4 = Distance(20)

print(f"\nInitial Distances:")
print(f"Distance 1: {distance1}")
print(f"Distance 2: {distance2}")
print(f"Distance 3: {distance3}")
print(f"Distance 4: {distance4}")

# Subtraction operator
print("\n--- Subtraction Examples ---")

result1 = distance1 - distance2
print(f"{distance1} - {distance2} = {result1}") # Expected: 64.5

result2 = distance3 - distance4
print(f"{distance3} - {distance4} = {result2}") # Expected: 100

result3 = distance2 - distance1
print(f"{distance2} - {distance1} = {result3}") # Expected: -64.5

# Verify types of results
print(f"\nType of result1: {type(result1)}")
```

Q4.Implement a Python program with two classes, one for distance in meters and another for distance in kilometers. Include functions to convert distances between these units.


```python
class Meters:
    def __init__(self, value):
        if not isinstance(value, (int, float)) or value < 0:
            raise ValueError("Meters value must be a non-negative number")
        self.value = value

    def to_kilometers(self):
        return self.value / 1000

    def __str__(self):
        return f"{self.value} meters"

    def __repr__(self):
        return f"Meters({self.value})"

class Kilometers:
    def __init__(self, value):
        if not isinstance(value, (int, float)) or value < 0:
            raise ValueError("Kilometers value must be a non-negative number")
        self.value = value

    def to_meters(self):
        return self.value * 1000

    def __str__(self):
        return f"{self.value} kilometers"

    def __repr__(self):
        return f"Kilometers({self.value})"


print("--- Demonstrating Distance Conversion Classes ---")

# 1. Create instances of Meters and Kilometers
m_distance = Meters(5500) # 5500 meters
k_distance = Kilometers(3.75) # 3.75 kilometers

print(f"\nInitial Distances:")
print(f"Original Meter Distance: {m_distance}")
print(f"Original Kilometer Distance: {k_distance}")

# 2. Demonstrate conversions
print("\n--- Conversion Example ---")

# Meters to Kilometers
meters_to_km = m_distance.to_kilometers()
print(f"{m_distance} is equal to {meters_to_km} kilometers")

# Kilometers to Meters
km_to_meters = k_distance.to_meters()
print(f"{k_distance} is equal to {km_to_meters} meters")
```
