Q1.
Write a Python function find_smallest_multiple(n: int) that uses a while loop to find the smallest positive integer x such that x is divisible by all numbers from 1 to n.

**Conditions:**

i. The function should keep incrementing x by 1 until it finds a number that satisfies the condition.

ii. Your solution should handle the edge case where n = 1 efficiently, returning 1 directly since 1 is divisible by itself.

```python
def find_smallest_multiple(n: int) -> int:
    if n == 1:
        return 1

    x = n
    while True:
        is_divisible_by_all = True
        for i in range(1, n + 1):
            if x % i != 0:
                is_divisible_by_all = False
                break
        if is_divisible_by_all:
            return x
        x += 1

print(f"Smallest multiple for n=1: {find_smallest_multiple(1)}")
print(f"Smallest multiple for n=3: {find_smallest_multiple(3)}")
print(f"Smallest multiple for n=10: {find_smallest_multiple(10)}")
```

### Question 2: Armstrong Numbers

Write a function `cubesum()` that accepts an integer and returns the sum of the cubes of individual digits of that number. Use this function to make functions `PrintArmstrong()` and `isArmstrong()` to print Armstrong numbers and to find whether a number is an Armstrong number.

```python
def cubesum(n: int) -> int:
    total_sum = 0
    num_str = str(abs(n))
    for digit_char in num_str:
        digit = int(digit_char)
        total_sum += digit ** 3
    return total_sum

def isArmstrong(n: int) -> bool:
    return n == cubesum(n)

def PrintArmstrong(limit: int):
    print(f"Armstrong numbers up to {limit}:")
    found_armstrong = False
    for i in range(1, limit + 1):
        if isArmstrong(i):
            print(i)
            found_armstrong = True
    if not found_armstrong:
        print(f"No Armstrong numbers found up to {limit}.")

print(f"cubesum(123): {cubesum(123)}")
print(f"cubesum(153): {cubesum(153)}")
print(f"cubesum(9): {cubesum(9)}")

print(f"isArmstrong(153): {isArmstrong(153)}")
print(f"isArmstrong(370): {isArmstrong(370)}")
print(f"isArmstrong(123): {isArmstrong(123)}")
```

Q3.
Why is operator precedence important? Give an example where neglecting precedence changes the result.

```python
result_correct_precedence = 5 + 2 * 3
print(f"Result with correct precedence (5 + (2 * 3)): {result_correct_precedence}")

result_forced_order = (5 + 2) * 3
print(f"Result with forced order ((5 + 2) * 3): {result_forced_order}")

print("\nAs you can see, the results are different. Python (and most programming languages)")
print("follows strict operator precedence rules to ensure consistent evaluation of expressions.")
```

Q6.
Write a program that takes a sentence as input and counts the frequency of vowels, consonants, digits, and special characters separately.

```python
def count_character_types(sentence: str):
    vowels = "aeiouAEIOU"
    digits = "0123456789"

    vowel_count = 0
    consonant_count = 0
    digit_count = 0
    special_char_count = 0

    for char in sentence:
        if char.isalpha():
            if char in vowels:
                vowel_count += 1
            else:
                consonant_count += 1
        elif char.isdigit():
            digit_count += 1
        elif not char.isspace(): # Count non-whitespace, non-alphanumeric as special
            special_char_count += 1

    print(f"Original Sentence: {sentence}")
    print(f"Vowels: {vowel_count}")
    print(f"Consonants: {consonant_count}")
    print(f"Digits: {digit_count}")
    print(f"Special Characters: {special_char_count}")

input_sentence = input("Enter a sentence: ")
count_character_types(input_sentence)

print("\n--- Example Test ---")
count_character_types("Hello World! 123 @#$")
```

Q7.
Write a Python program to create a dictionary of studentsâ€™ names as keys and their marks as values.
Then:
A. Print the student with the highest marks
B. Print the student with the lowest marks

```python
student_marks = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78,
    "David": 95,
    "Eve": 88,
    "Frank": 70
}

print("Student Marks:", student_marks)

if student_marks:
    highest_mark_student = max(student_marks, key=student_marks.get)
    highest_mark = student_marks[highest_mark_student]
    print(f"\nStudent with the highest marks: {highest_mark_student} ({highest_mark})")
else:
    print("\nNo students in the dictionary to find the highest marks.")

if student_marks:
    lowest_mark_student = min(student_marks, key=student_marks.get)
    lowest_mark = student_marks[lowest_mark_student]
    print(f"Student with the lowest marks: {lowest_mark_student} ({lowest_mark})")
else:
    print("No students in the dictionary to find the lowest marks.")
```
